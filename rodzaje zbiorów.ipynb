# Wstęp

```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
sns.set()

np.random.seed(42)
np.set_printoptions(edgeitems=30, linewidth=100000,
                    formatter=dict(float=lambda x: f'{x:.2f}'))
```

```python
from sklearn.datasets import load_iris

raw_data = load_iris()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
print(raw_data_copy['DESCR']) # opis danych
```

# Przygotowanie danych

```python
data = raw_data_copy['data']
target = raw_data_copy['target']

print(data)
print(target)
```

```python
# połączenie danych z targetem
all_data = np.c_[data, target]
print(all_data[:5])
```

```python
df = pd.DataFrame(all_data,
                  columns=raw_data_copy['feature_names'] + ['target'])
df.head()
```

```python
df.info() # nie ma braków danych
```

```python
df.describe().T
```

```python
df.target.value_counts()
# tyle samo różnych klas
```

```python
data = df.copy()
target = data.pop('target')
data.head()
```

```python
target.head()
```

# Podział na zbiór treningowy i testowy - iris data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data, target, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

# 112 danych wierszy testowych, 38 treningowych
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data, target, random_state=42, test_size=0.1, stratify=target)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(y_train.value_counts())
print(y_test.value_counts())

# gdy stratify = target dzieli zmienną docelową na równe części
```

# Breast cancer data

```python
from sklearn.datasets import load_breast_cancer

raw_data = load_breast_cancer()
raw_data_copy = raw_data.copy()
raw_data_copy.keys()
```

```python
print(raw_data_copy['DESCR'])
```

```python
data = raw_data_copy['data']
target = raw_data_copy['target']

print(data[:5])
print(target[:5])
```

```python
all_data = np.c_[data, target]
print(all_data[:5])
```

```python
df = pd.DataFrame(data=all_data,
                  columns=list(raw_data['feature_names']) + ['target'])
df.head()
```

```python
df.target.value_counts() # nie równy podział klas
```

```python
data = df.copy()
target = data.pop('target')
data.head()
# oddzelamy dane i target
```

```python
target.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data, target, random_state=42)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(y_train.value_counts())
print(y_test.value_counts())

#klasa 1 przewyższa klase 0 liczebnością
```
