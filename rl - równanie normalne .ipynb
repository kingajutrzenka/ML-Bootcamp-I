# Wstęp

```python
import numpy as np
```

```python
X1 = np.array([1,2,3,4,5,6])
Y = np.array([3000,3250,3500,3750,4000,4250])
m = len(X1)
print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba próbek: {m}')
```

# Regresja liczona ręcznie

```python
#zmieniamy rozmiar
X1 = X1.reshape((m,1))
print(X1)
print(X1.shape)
```

```python
bias = np.ones((m,1))
print(bias)
print(bias.shape)
```

```python
X = np.append(bias, X1, axis=1)
print(X)
print(X.shape)
```

**Korzystamy z tego ostatniego W**

```python
# iloczyn macierzy X transponowane * X
np.dot(X.T, X)
```

```python
# odwracamy tą macierz
L = np.linalg.inv(np.dot(X.T, X))
L
```

```python
# liczymy drugi nawias
P = np.dot(X.T, Y)
```

```python
# Mnożymy i gotowe
np.dot(L, P)
```

wyszło nam 2750 i 250. 2750 to w0 (b), 250 to w1 (a), czyli postać modelu to:
**Y = 2750 + 250X1**

# Regresja scikit - learn

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(X1, Y)

print(regression.intercept_) # nasze w0
print(regression.coef_[0]) # nasze w1
```

też wychodzi **Y = w0 + w1x1
