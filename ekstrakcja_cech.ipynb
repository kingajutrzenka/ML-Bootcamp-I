# Za≈Çadowanie danych

```python
def fetch_financial_data(company='AMZN'):
    """
    This function fetches stock market quotations.
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df_raw = fetch_financial_data()
df_raw.head()
```

```python
df = df_raw.copy()
df = df[:5]
df.info()
```

# Nowe zmienne

```python
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df
```

# Dyskretyzacja zmiennej ciƒÖg≈Çej

```python
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=3) # samo dzieli
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195))
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195), labels=['small', 'medium', 'high'])
df
```

```python
pd.get_dummies(df, drop_first=True, prefix='height', dtype = 'int')
```

# Ekstrakcja cech

```python
df = pd.DataFrame(data={'lang': [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df
```

```python
df['lang_number'] = df['lang'].apply(len) # liczy ile jest jƒôzyk√≥w
df
```

```python
df['PL_flag'] = df['lang'].apply(lambda x: 1 if 'PL' in x else 0) # funkcja sprawdza, czy jest polska
df
```
